# Importar Librerias

```python
%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

# Ecuaciones de Lorenz

```python
def lorenz(p, a, b, c, tfin, h, tinit=0):
    xs = [p[0]]
    ys = [p[1]]
    zs = [p[2]]
    for i in np.arange(tinit, tfin, h):
        v = np.array([
            a*(p[1]-p[0]),
            p[0]*b-p[2]*p[0]-p[1],
            p[0]*p[1]-c*p[2]
        ])
        p = p+ h*v
        xs.append(p[0])
        ys.append(p[1])
        zs.append(p[2])
    return xs, ys, zs
```

# Ejecución

```python
p = np.array([0, 1, 1.05])
xs, ys, zs = lorenz(p, 38, 38, 14, 100, 0.01, tinit=0)
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot(xs, ys, zs)
```
