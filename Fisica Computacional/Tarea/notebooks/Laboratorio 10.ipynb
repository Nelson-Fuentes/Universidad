# Función RK4

```python
def rk4(function, t, tfin, v, x, h, m):
    pt = [t]
    pv = [v]
    px = [x]

    while t<tfin:
        for i in range(m):
            a  = eval(function)
            k1 = h*a            

            x1 = x
            v1 = v
            t1 = t

            t = t1+0.5*h
            x = x1+h*0.5*v1
            v = v1+0.5*k1

            a  = eval(function)
            k2 = h*a

            t = t1+0.5*h
            x = x1+0.5*h*(v1+0.5*k1)
            v = v1+0.5*k2

            a  = eval(function)
            k3 = h*a

            t = t1+h
            x = x1+h*v1+h*k2*0.5
            v = v1+k3

            a  = eval(function)
            k4 = h*a;

            x = x1+h*v1+h*(k1+k2+k3)/6
            v = v1+(k1+2*k2+2*k3+k4)/6
            t = t1+h
            if x>math.pi:
                x=x-2*math.pi
            if x<-math.pi:
                x=x+2*math.pi
        px.append(x)
        pv.append(v)
    return px, pv
```

# Ejercicios
## Escriba un código para resolver el problema de un péndulo simple aplicando las secciones de Poincaré.

$\frac{d^{2}\theta}{dt^{2}} = -\frac{g}{l}\theta$

```python
m=20
g = 1
l = 1
tfin = 10000
x=1
v=-1
t=0
h=2*math.pi/(1.7*m)
function='-g/l'
px, pv = rk4(function, t, tfin, v, x, h, m)
fig, axis = plt.subplots(constrained_layout=True)
axis.scatter(px,pv,marker='o')
axis.set_aspect(4000)
axis.set(xlabel='Desplazamiento', ylabel='Velocidad')
axis.grid()
```
