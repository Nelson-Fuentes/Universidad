# Metodo que retorna los puntos de Lissajous

```python
def Lissajous(lnm, m, v, p, h, tf):
    k = m * (lnm**2)
    a = -k * p / m
    ps = [ p ]
    for t in np.arange(0, tfin, h):
        a = -k * p / m
        v = v + a*h
        p  = p  + v*h
        ps.append(p)
    return ps
```

#  Con estos datos dibujar las figuras de Lissajous con las condiciones iniciales

(a) x = 1, vx = 0, y = +1, vy = 0

```python
p = np.array([ 1, 1 ])
v = np.array([ 0, 0 ])

m = np.array([ 1, 1]) 
ln = np.array([ 1 , 1 ])
h = 0.01
tfin=100
ps = Lissajous(ln, m,v,p,h, tfin)
fig, ax = pt.subplots()
ax.plot([p[0] for p in ps], [p[1] for p in ps])
ax.grid()
ax.set_aspect('equal')
```

# Con estos datos dibujar una figura de Lissajous con las condiciones iniciales en 3D como referencia.
